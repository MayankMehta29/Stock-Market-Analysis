# Time Series Data
A time series is a sequence of numerical data points taken at successive equally spaced points in time. <br>
In investing, a time series tracks the movement of stock price, over a specified period of time.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import matplotlib.dates as mdates
```

```python
plt.plot(ibm.index, ibm['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
#Subplots
f, ax = plt.subplots(3, 2, figsize=(10,15), sharex=True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
f.gca().xaxis.set_major_locator(mdates.YearLocator())

ax[0,0].plot(nvda.index, nvda['Adj Close'], color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('NVIDIA');

ax[0,1].plot(intc.index, intc['Adj Close'], color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('INTEL');

ax[1,0].plot(qcom.index, qcom['Adj Close'], color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QUALCOMM');

ax[1,1].plot(amd.index, amd['Adj Close'], color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('AMD');

ax[2,0].plot(msft.index, msft['Adj Close'], color='m')
ax[2,0].grid(True)
ax[2,0].tick_params(labelrotation=90)
ax[2,0].set_title('MICROSOFT');

ax[2,1].plot(aapl.index, aapl['Adj Close'], color='darkorange')
ax[2,1].grid(True)
ax[2,1].tick_params(labelrotation=90)
ax[2,1].set_title('APPLE');


```

# Step 3: Moving Windows

### Daily Percentages

```python
daily_close_ibm = ibm[['Adj Close']]

# Daily returns
daily_pct_change_ibm = daily_close_ibm.pct_change()

# Replace NA values with 0
daily_pct_change_ibm.fillna(0, inplace=True)

daily_pct_change_ibm.head()
```

```python
daily_pct_change_ibm.hist(bins=50)

# Show the plot
plt.show()
```

```python
daily_close_nvda = nvda[['Adj Close']]

# Daily returns
daily_pct_change_nvda = daily_close_nvda.pct_change()

# Replace NA values with 0
daily_pct_change_nvda.fillna(0, inplace=True)

daily_close_intc = intc[['Adj Close']]

# Daily returns
daily_pct_change_intc = daily_close_intc.pct_change()

# Replace NA values with 0
daily_pct_change_intc.fillna(0, inplace=True)

daily_close_qcom = qcom[['Adj Close']]

# Daily returns
daily_pct_change_qcom = daily_close_qcom.pct_change()

# Replace NA values with 0
daily_pct_change_qcom.fillna(0, inplace=True)

daily_close_amd = amd[['Adj Close']]

# Daily returns
daily_pct_change_amd = daily_close_amd.pct_change()

# Replace NA values with 0
daily_pct_change_amd.fillna(0, inplace=True)

daily_pct_change_amd.head()

daily_close_msft = msft[['Adj Close']]

# Daily returns
daily_pct_change_msft = daily_close_msft.pct_change()

# Replace NA values with 0
daily_pct_change_msft.fillna(0, inplace=True)

daily_close_aapl = aapl[['Adj Close']]

# Daily returns
daily_pct_change_aapl = daily_close_aapl.pct_change()

# Replace NA values with 0
daily_pct_change_aapl.fillna(0, inplace=True)

```

```python
import seaborn as sns
sns.set()
```

```python
import seaborn as sns
# Set up the matplotlib figure
f, axes = plt.subplots(3, 2, figsize=(15,10))

# Plot a simple histogram with binsize determined automatically
sns.distplot(daily_pct_change_nvda['Adj Close'], color="r", ax=axes[0, 0], axlabel='NVIDIA');

# Plot a kernel density estimate and rug plot
sns.distplot(daily_pct_change_intc['Adj Close'], color="g", ax=axes[0, 1], axlabel='INTEL');

# Plot a filled kernel density estimate
sns.distplot(daily_pct_change_qcom['Adj Close'], color="b", ax=axes[1, 0], axlabel='QUALCOMM');

# Plot a historgram and kernel density estimate
sns.distplot(daily_pct_change_amd['Adj Close'], color="y", ax=axes[1, 1], axlabel='AMD');


# Plot a historgram and kernel density estimate
sns.distplot(daily_pct_change_msft['Adj Close'], color="m", ax=axes[2, 0], axlabel='MICROSOFT');


# Plot a historgram and kernel density estimate
sns.distplot(daily_pct_change_aapl['Adj Close'], color="darkorange", ax=axes[2, 1], axlabel='APPLE');
```

# Step 4: Volatality

```python
import numpy as np
```

```python
min_periods = 75

# Calculate the volatility
vol = daily_pct_change_ibm.rolling(min_periods).std() * np.sqrt(min_periods) 

vol.fillna(0,inplace=True)

vol.tail()
```

```python
# Plot the volatility
vol.plot(figsize=(10, 8))

# Show the plot
plt.show()
```

# Rolling Means (Trends and Seasonality)

```python

ibm_adj_close_px = ibm['Adj Close']
# Short moving window rolling mean
ibm['50'] = ibm_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
ibm['200'] = ibm_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ibm[['Adj Close','50', '200']].plot(title="IBM")

# Show plot
plt.show()

nvda_adj_close_px = nvda['Adj Close']
# Short moving window rolling mean
nvda['50'] = nvda_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
nvda['200'] = nvda_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
nvda[['Adj Close', '50', '200']].plot(title="NVIDIA")

# Show plot
plt.show()

qcom_adj_close_px = qcom['Adj Close']
# Short moving window rolling mean
qcom['50'] = qcom_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
qcom['200'] = qcom_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
qcom[['Adj Close', '50', '200']].plot(title="QUALCOMM")

# Show plot
plt.show()

amd_adj_close_px = amd['Adj Close']
# Short moving window rolling mean
amd['50'] = amd_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
amd['200'] = amd_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
amd[['Adj Close', '50', '200']].plot(title="AMD")

# Show plot
plt.show()

aapl_adj_close_px = aapl['Adj Close']
# Short moving window rolling mean
aapl['50'] = aapl_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
aapl['200'] = aapl_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
aapl[['Adj Close', '50', '200']].plot(title="APPLE")

# Show plot
plt.show()

msft_adj_close_px = msft['Adj Close']
# Short moving window rolling mean
msft['50'] = msft_adj_close_px.rolling(window=50).mean()

# Long moving window rolling mean
msft['200'] = msft_adj_close_px.rolling(window=200).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
msft[['Adj Close', '50', '200']].plot(title="Microsoft")

# Show plot
plt.show()
```
